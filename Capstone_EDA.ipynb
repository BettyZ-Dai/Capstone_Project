```python
df = df.drop(['DATE','PAGE_VIEWS','TIME_ON_SITE_TOTAL','VISITOR_BOUNCE','VISITS'], axis=1)
```

```python
df['MEDIA_COST'].fillna(0, inplace=True)
```

```python
df[(df['IMPRESSIONS'].isnull()) & (df['CLICKS']==0)]
```

```python
df['IMPRESSIONS'].fillna(0, inplace=True)
```

```python
df[(df['_POSITION'].isnull()) & (df['CLICKS']!=0)]
```

```python
df['_POSITION'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['CTR'] = df['CLICKS']/df['IMPRESSIONS']
```

```python
df['CTR'].isnull().sum()
```

```python
df['CTR'].fillna(0, inplace=True)
```

#### Add categorical label

The average click-through rate on paid search ads is about 2%; 
The Bing Ads benchmark average CTR in health is 2.90% for paid search; 
The Google Ads benchmark average CTR in health is 3.27%-3.82% for paid search  (https://www.wordstream.com/blog/ws/2016/02/29/google-adwords-industry-benchmarks)

```python
df.loc[df['CTR']>=0.03,'Flag'] = 1
df.loc[df['CTR']<0.03,'Flag'] = 0
```

```python
df['Flag'].isnull().sum()
```

```python
df.to_csv('V1.csv',index_label=False)
```

#### Feature selection

```python
scaler = StandardScaler()
X = df_new.values
X_new = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=0.95)
pca.fit(X_new)
m = pca.transform(X_new)
```

```python
pca.explained_variance_ratio_
```

```python
import numpy as np
most_important = [np.abs(pca.components_[i]).argmax() for i in range(pca.components_.shape[0])]
```

```python
feature_names = df_new.columns
```

```python
most_important_names = [feature_names[most_important[i]] for i in range(pca.components_.shape[0])]
```

```python
most_important_names
```

```python
xvector = pca.components_[0]
yvector = pca.components_[1]
```

```python
xs = m[:,0]
ys = m[:,1]
```

```python
import matplotlib.pyplot as plt 

for i in range(len(xvector)):
# arrows project features as vectors onto PC axes
    plt.arrow(0, 0, xvector[i], yvector[i], color='k', alpha=0.5)
    plt.text(xvector[i]*1.15, yvector[i]*1.15, feature_names[i], color='g')

plt.xlim(-0.1,0.8)
plt.ylim(-0.8,0.9)

plt.show()
```
